##Task-1  

I have used principles of proportionality and geometric similarity (assume that black bears are geometrically similar to each other) to derive a reasonable model relating their length to their weight.  

I have used methods demonstrated in class to graphically derive coefficients for my model and test the fit of the model to the data as follows


- Created a scatterplot for the original data


```
import numpy as np
import matplotlib.pyplot as plt
W = np.array([27,70,99,47,50,34,47,41,34,52,16,99,15,41,36,81,35,54,45,45])
L = np.array([351,422,457,329,381,335,376,356,348,378,259,439,265,351,367,417,328,401,381,361])
plt.scatter(W,L)
plt.ylabel('length (cm)')
plt.xlabel('weight (kg)')
plt.title ('scatterplot for origial data')
plt.grid()
plt.show()
```

after review scatterplot as we discussed in class I tried to apply same method for this model. I assume that given length to their weight may have power relation. 

I decide to find coefficient for following model using ruler on scatter plot.

$$L = aW^b$$

```
plt.scatter(np.log(W), np.log(L))
plt.ylim(4.00,6.2)
plt.xlim(0.00, 5)
plt.ylabel('ln L')
plt.xlabel('ln W')
plt.title ('scatterplot for the origial data')
plt.grid()
plt.show()
```

#Graphically Estimating coefficients

Useing ruler on my computer screen and estimate that coefficent as as followings 
$$ ln L = ln(aW^b) = ln a + b ln W$$

$$ ln W = 0, ln a = ln L \approx 4.87,  slop = b \approx 0.27$$

$ ln L = ln a + b ln W$

$ ln a = 4.87$

$ b = 0.27 $


```
plt.scatter(np.log(W), np.log(L))
lna=4.87
b=0.27
x = np.linspace(2,5)
y = lna + b * x
plt.plot (x,y)
plt.ylabel('ln L')
plt.xlabel('ln W')
plt.grid()
plt.show()
```

##Summerize model and plot angainst data##

$ln a = 4.87 \Longrightarrow a = e^{4.87} = 130.32  $

$ b = 0.27 $

$L = aW^b \Longrightarrow L = 130.32W^{0.27}$

my model is $L = 130.32W^{0.27}$

```
a=130.32
b=0.27
plt.xlim(0.0,105.0)
plt.ylim(0.00, 500.00)
x = np.linspace(0,105,105)
y = a * x ** b
plt.scatter(W,L)
plt.plot (x,y, label = '$L = 130.32W^{0.27}$')
plt.xlabel("$W$")
plt.ylabel("$L$")
plt.title ("Black Bear length to weight\nOriginal data with curve that reprsent model")
plt.legend()
plt.show()
```

computation of upper and lower bounds for the Chebyshev criteria, as presented in class and Section 3.2 of your text.


```
import math as m
y = a * x ** b
upperbond = 0
for i,j in zip(W,L):
  current_chebyshev = np.abs(y[i]-j)
  if current_chebyshev > upperbond:
    upperbond = current_chebyshev
    lowerbond = m.sqrt((y[i]-j)**2/len(L))
print(f'chebyshev critaria for model L = aW^b  UppberBond is {upperbond:.2f} and LowerBond is {lowerbond:.2f}')
```
